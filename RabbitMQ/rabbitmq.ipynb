```python
RabbitMQ provides streaming, queuing & pub-sub capabilities
Implementation of the protocol - Advanced Messqge Queue Protocol

It makes a system scalable, decoupled and performant
```

```python
Producer -> Exchange

Exchange can be connected to any no. of queues based on strategy like fan-out or topics

Exchange is connected to the queues through bindings
Queues are connected to the consumers who are their subscribers
```

```python
Fanout -> It sends to all the consumers available

Direct -> Routing key matches binding key -> Message is sent to that queue

Topic -> Partial match between binding key and routing key

Header -> Sent according to the header

Default -> routing key = queue name
```

```python
Application and message metadata define how it will flow through the broker mechanism
```

```python
RabbitMQ has good security and management tools
```
