| Structure      | Lookup   | Insert   | Delete   | Ordered?                     |
| -------------- | -------- | -------- | -------- | ---------------------------- |
| Array/List     | O(1)     | O(n)     | O(n)     | Yes (index order)            |
| Linked List    | O(n)     | O(1) [at head] O(n) [at tail]     | O(1)  [at head] O(n) [at tail]  | Yes (sequence, unsorted)               |
| Stack          | O(n)     | O(1) [at head]    | O(1) [at head]     | LIFO                         |
| Queue          | O(n)     | O(1)     | O(1)     | FIFO                         |
| HashMap        | O(1)     | O(1)     | O(1)     | No                           |
| BST (balanced) | O(log n) | O(log n) | O(log n) | Yes                          |
| Heap           | O(1)*    | O(log n) | O(log n) | Partially (by heap property) |
| Trie           | O(L)     | O(L)     | O(L)     | By prefix                    |
| Graph          | O(V+E)   | O(1)     | O(1)     | —                            |

| **Goal / Operation Pattern**                       | **Best Data Structure**                             | **Why**                                           |
| -------------------------------------------------- | --------------------------------------------------- | ------------------------------------------------- |
| Fast lookup by key                                 | **HashMap / Dictionary**                            | O(1) average lookup and insertion                 |
| Maintain elements in sorted order                  | **Balanced BST (e.g., Red-Black Tree)** or **Heap** | Ordered traversal or min/max extraction           |
| Constant-time min/max retrieval                    | **Heap (Priority Queue)**                           | `O(1)` for min/max; `O(log n)` for insert/delete  |
| Access elements by index                           | **Array / List**                                    | O(1) random access                                |
| Frequent insert/delete anywhere                    | **Linked List**                                     | O(1) insertion/deletion (if node reference known) |
| Need both fast access and order                    | **OrderedDict / TreeMap**                           | Keeps insertion or key order                      |
| Track unique elements                              | **Set / HashSet**                                   | Enforces uniqueness automatically                 |
| Sequence of actions (undo/redo, recursion)         | **Stack**                                           | LIFO behavior fits call history                   |
| Scheduling or queues                               | **Queue / Deque**                                   | FIFO behavior for tasks or messages               |
| Prefix or autocomplete                             | **Trie**                                            | Efficient prefix matching                         |
| Hierarchical data (file system, ML decision trees) | **Tree**                                            | Parent-child representation                       |
| Network, dependency, or path-finding problems      | **Graph (Adjacency List / Matrix)**                 | Represents connections                            |
| Sliding window / LRU cache                         | **Deque + HashMap / OrderedDict**                   | Constant-time eviction                            |
| Range sum or min/max queries                       | **Segment Tree / Fenwick Tree**                     | Fast updates and range queries                    |

# Linked list

```python
class Node:
    def __init__(self, data):
        self.data = data     # store the data
        self.next = None     # pointer to next node

class LinkedList:

    def __init__(self):
        self.head = None   # initially empty list

    # Insert at the end - O(n)
    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        current = self.head
        while current.next: # This operation is O(n)
            current = current.next
        current.next = new_node

    # Insert at the beginning - O(1)
    def prepend(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    # Delete by value
    def delete(self, data):
        current = self.head

        # If list is empty
        if not current:
            return

        # If head needs to be removed
        if current.data == data:
            self.head = current.next
            return

        # Search for the node to delete
        while current.next and current.next.data != data:
            current = current.next

        # If found, unlink it
        if current.next:
            current.next = current.next.next

    def find(self, data):
        current = self.head
        while current:
            if current.data == data:
                return True
            current = current.next
        return False

    # Print all elements
    def display(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")
```

```python
ll = LinkedList()
ll.append(1)
ll.append(2)
ll.prepend(0)
ll.append(53)
ll.append(4)
import numpy as np

for i in np.arange(40):
    ll.append(np.random.random(1))
%time ll.append(43)

for i in np.arange(4000):
    ll.append(np.random.random(1))
%time ll.append(443) # Takes longer to insert, as n has increased


print(ll.find(53))
ll.delete(2)
ll.display()
```

# Binary Search Tree (BST)

THink of binary search, Each node has atmost 2 children, with the left always being smaller and the right always being greater.

The more balanced your tree, the closer to logarithmic time your operations will be.
The more skewed it becomes, the closer to linear time it gets, as it tends to being a linked-list.

```python
class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None
        self.parent = None
```

```python
class BST:

    def __init__(self):
        self.root = None

    def insert(self, key):
        if self.root is None:
            self.root = Node(key)
        else:
            self._insert(self.root, key)

    def _insert(self, current, key):
        if key < current.key:
            if current.left is None:
                current.left = Node(key)
            else:
                self._insert(current.left, key)
        elif key > current.key:
            if current.right is None:
                current.right = Node(key)
            else:
                self._insert(current.right, key)

    def search(self, key):
        return self._search(self.root, key)

    def _search(self, current, key):
        if current is None:
            return False
        if key == current.key:
            return True
        elif key < current.key:
            return self._search(current.left, key)
        else:
            return self._search(current.right, key)

    def inorder(self):
        result = []
        self._inorder(self.root, result)
        return result

    def _inorder(self, node, result):
        if node:
            self._inorder(node.left, result)
            result.append(node.key)
            self._inorder(node.right, result)
        else:
            return result

    def delete(self, key):
        self.root = self._delete(self.root, key)

    def _delete(self, node, key):
        if node is None:
            return node

        # Step 1: Find the node
        if key < node.key:
            node.left = self._delete(node.left, key)
        elif key > node.key:
            node.right = self._delete(node.right, key)
        else:
            # Step 2: Node found — handle the 3 cases

            # Case 1: No child
            if node.left is None and node.right is None:
                return None

            # Case 2: One child
            if node.left is None:
                return node.right
            elif node.right is None:
                return node.left

            # Case 3: Two children
            # Find inorder successor (smallest in right subtree)
            min_node = self._min_value_node(node.right)
            node.key = min_node.key

            # Delete the inorder successor
            node.right = self._delete(node.right, min_node.key)

        return node

    def _min_value_node(self, node):
        current = node
        while current.left is not None:
            current = current.left
        return current
```

```python
bst = BST()

# Insert some values
for val in [50, 30, 70, 20, 40, 60, 80]:
    bst.insert(val)

# Print in sorted order
print("Inorder traversal:", bst.inorder())
# → [20, 30, 40, 50, 60, 70, 80]

# Search for elements
print("Search 60:", bst.search(60))   # True
print("Search 25:", bst.search(25))   # False

bst.delete(40)
print("Inorder after deleting 20:", bst.inorder())
```
