```python
import uuid
from pydantic import BaseModel

class OrderItem(BaseModel):
    key: str
    name: str
    price: float
    quantity: int = 1

random_items = [
    OrderItem(key=str(uuid.uuid4()), name="Wireless Mouse", price=799.0),
    OrderItem(key=str(uuid.uuid4()), name="Monitor", price=17999.0),
    OrderItem(key=str(uuid.uuid4()), name="Chair", price=6999.0, quantity=2),
    OrderItem(key=str(uuid.uuid4()), name="USB-C Cable", price=299.0, quantity=2),
]
```

```python
import numpy as np
num_items=3
items=np.random.choice(random_items, num_items)
items
```

```python
[a.name for a in items]
```
