# Introduction

Redis is an in-memory, single-threaded key-value datastore

 - In-memory -> It has sub milli-second performance
 - Single threaded -> Queues operations
 - Key-value datastores -> It can store multiple data types around redis core KV store

- It supports AOF(Append only File) and Snapshhot based backups
- It follows the main node-secondary node strategy
- When load increases, we can increase the no. of replica sets, no. of shards

Redis hot key issue - When keys are spread across diff. replicas, but one key is very hot(highly accessed), that replica may get overloaded, in that case, you add a number at the end of that key and make it accessible across more replicas, thereby spreading the load

Expiration policy in cache - Different strategies are there, we used TTL of 60s; LRU strategy is also possible 

Using as a ratelimiter for an expensive service

Redis stream for distributing work among different workers in a distributed system

Redis pub/sub -> Agent registry refresh

```python
import redis

pool = redis.ConnectionPool(host='localhost', port=6379, db=0)
r = redis.Redis(connection_pool=pool)

# Check specific keys
r.keys('*')

#------------
# Check number of keys in database
r.dbsize()

#------------
# set key value
r.set('key', 'value')
r.set('key', 'value', ex=10, nx=True)

#------------
# get value by key
value = r.get('key')

#------------
# syntax : delete keys
r.delete('key')
r.delete('key1', 'key2', 'key3')

#------------
# Check if key exists
r.exists('key')

#------------
# set expiry to key
expireInSeconds = 30
r.expire('key', expireInSeconds)

#------------
# remove expiry from key
r.persist('key')

#------------
# find (remaining) time to live of a key
r.ttl('key')

#------------
# increment a number
r.incr('key')

#------------
# decrement a number
r.decr('key')

#------------
# use the method below to execute commands directly
r.execute_command('SET', 'key', 'value')
```
