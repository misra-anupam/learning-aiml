# Sliding window

Problem: Find the smallest subarray with sum ≥ target.

Idea:

Expand the window by moving the right pointer until the sum ≥ target

Then shrink it from the left to make it as small as possible

```python
import numpy as np

arr = np.random.random_integers(low=1,high=45, size=15).tolist()
target = 44
arr
```

```python
def sliding_min(arr, target):
    low = 0
    high = 1
    min_len = None

    while high<len(arr):

        arr_sum = sum(arr[low:high])

        while arr_sum<=target and high<len(arr):
            high += 1
            arr_sum = sum(arr[low:high])

        if min_len is None:
            min_len=high-low
        elif (high-low)<min_len:
                min_len=high-low

        low += 1
        high = low+1

    return min_len
```

```python
sliding_min([2, 1, 5, 2, 3, 2], 5)
```

# Two pointer technique

The two-pointer technique uses two indices (pointers) that traverse an array (or string) from different directions or at different speeds to solve problems efficiently.

You avoid nested loops (O(n²)) and bring most problems down to O(n) time.

You can often apply two pointers when:

* The array or string is sorted (or can be sorted).
* You need to find pairs or triplets that meet a condition.
* You need to remove, reverse, partition, or merge elements in-place.
* You need to compare or move elements relative to each other.

Problem: Remove duplicates so that each element appears only once.

```python
arr = [0,1,1,1,3,4,5,5,6,6,7,7,7,7,8,9,10,11,11,12,13,14]

def remove_duplicates(arr):

    i = len(arr)-1
    j = i-1

    while j>=0:

        while (arr[j] == arr[i]) and (j>=0):
            arr.pop(j)
            j-=1
        i -= 1
        j = i-1
    return arr
```

```python
remove_duplicates(arr)
```

Reverse an Array In-Place

```python
arr = np.arange(20)
print(arr)
def reverse_in_place(arr):
    low = 0
    high = len(arr)-1

    while high>low:

        arr[low] , arr[high] = arr[high] , arr[low]
        low = low + 1
        high = high - 1
    return arr

print(reverse_in_place(arr))
```

## Fast-slow pointers: 
Two pointers move at different speeds; when they meet, there’s a cycle or a pattern.

| Move zeroes                 | Same direction        |
| Merge two sorted arrays     | Two arrays            |

# Prefix sum / cumulative sum

Normally, computing the sum of each subarray takes O(n) time.
With prefix sums, you preprocess once in O(n) and answer each query in O(1).

This is especially useful for:

* Range sum queries

* Finding equilibrium indices

* Problems involving cumulative quantities (rainwater, differences, etc.)

```python
# Equilibrium index


l = np.random.random_integers(low=0, high=10, size=40)
print(l)
```

```python
presum = [l[0]]
postsum = [l[len(l)-1]]
pre=1
post=len(l)-2

while pre<len(l) and post>0:

    presum.append(presum[pre-1]+l[pre])
    postsum.append(postsum[pre-1]+l[post])

    pre+=1
    post-=1
```
